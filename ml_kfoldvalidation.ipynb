```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# 1. CSV 파일 불러오기
df = pd.read_csv('concrete_data.csv')

# 2. 데이터 확인 (특징과 목표 변수 선택)
X = df.drop('concrete_compressive_strength', axis=1)
y = df['concrete_compressive_strength']

# 3. 1차 다항식 회귀모델 설정
model = LinearRegression()

# 4. 5-Fold Cross-Validation 실행
scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')

# MSE 값은 음수로 반환되므로, 이를 양수로 변환
mse_scores = -scores

# 평균 MSE 계산
mean_mse = np.mean(mse_scores)
std_mse = np.std(mse_scores)

print(f"Mean MSE: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python

from sklearn.pipeline import make_pipeline

# 2. 목표 변수와 특징 선택
X = df.drop('concrete_compressive_strength', axis=1)  # 목표 변수 'concrete_compressive_strength'
y = df['concrete_compressive_strength']

# 3. 다항식 회귀 모델 평가
results = {}

for degree in range(2, 6):  # 2차부터 5차까지
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

    # 4. 5-Fold Cross-Validation 실행
    scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')

    # MSE 값은 음수로 반환되므로, 이를 양수로 변환
    mse_scores = -scores

    # 평균 MSE 계산
    mean_mse = np.mean(mse_scores)
    std_mse = np.std(mse_scores)

    # 결과 저장
    results[degree] = {'Mean MSE': mean_mse, 'Std MSE': std_mse}

# 결과 출력
for degree, metrics in results.items():
    print(f"Degree: {degree} - Mean MSE: {metrics['Mean MSE']}, Standard Deviation of MSE: {metrics['Std MSE']}")
```
