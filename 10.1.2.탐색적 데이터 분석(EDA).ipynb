```python
# 필요한 패키지 설치

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
%matplotlib inline
```

```python
# 데이터 불러오기
# https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand
df = pd.read_csv("/content/hotel_bookings.csv")

# 데이터 샘플 확인
df.head()
#df.tail()
```

```python
# 각 컬럼의 통계치 확인
df.describe()
```

```python
numeric_df = df.drop(columns=df.select_dtypes(include=['object']).columns)
numeric_df.skew()
```

```python
df.skew(numeric_only=True)
```

```python
# 각 컬럼의 첨도 확인

# 숫자형 데이터만 선택하여 첨도 계산
df.select_dtypes(include=['number']).kurtosis()
```

```python
# 특정 변수 분포 시각화
plt.rcParams['figure.dpi'] = 300 #Runtime Configuration Parameters, 줄여서 rcParams
sns.distplot(df['lead_time'])
```
